```python
# n = "123402"
n= "7755"

length = len(n) 
summary = 0


# 왼쪽 부분의 자릿수 합 더하기
for i in range(length // 2):
    summary += int(n[i])


# 오른쪽 부분의 자릿수 합 빼기
for i in range(length // 2, length):
    summary -= int(n[i])


# 왼쪽 부분과 오른쪽 부분의 자릿수 합이 동일한지 검사
if summary == 0:
    print("LUCKY")
else:
    print("READY")
```
