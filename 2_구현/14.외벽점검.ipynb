```python

import  itertools
def solution(n,weak,dist):
    length = len(weak)  
    
    ## 1. weak  2 
    for i in range(length):
        weak.append(weak[i] + n)
        
    ## 2.      1   
    answer = len(dist) + 1
    
    ## 3. 0 length-1 start     
    for start in range(length):
        for friends in list(itertools.permutations(dist , len(dist))):
            cnt = 1  ##  
        
            ##     
            position = weak[start] + friends[cnt - 1]

            ##   
            for i in range(start , start + length):
                if position < weak[i]:
                    cnt+=1
                    if cnt > len(dist):
                        break
                    position = weak[i] + friends[cnt-1]

            answer = min(cnt,answer)

    if answer > len(dist):
        return -1
    
    return answer
        
```
