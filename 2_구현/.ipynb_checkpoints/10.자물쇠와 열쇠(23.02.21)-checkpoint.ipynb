```python
import numpy as np

lock = [[1,1,1],[1,1,0],[1,0,1]]
key = [[0,0,0],[1,0,0],[0,1,1]]

ans = [[1,1,1],[1,1,1],[1,1,1]]
```

```python
## lock 
def expand(lock,n): #
    E_lock=np.pad(lock, (n, n),
       mode='constant',
       constant_values=0)
    
    return E_lock

expand(lock,2)
```

```python
## 90 !!!
def rotate(key):
    n = len(key) #   
    m = len(key[0]) #   
    new = [[0] * n for _ in range(m)]
    for i in range(n):
        for j in range(m):
            new[j][n-i-1] = key[i][j]
    return new
```

```python
# key 90  4 
def keys(key):
    a=key
    b=rotate(key)
    c=rotate(rotate(key))
    d=rotate(rotate(rotate(key)))
    key_arr = [a,b,c,d]
    return key_arr

for i in keys(key):
    print(i)
```
