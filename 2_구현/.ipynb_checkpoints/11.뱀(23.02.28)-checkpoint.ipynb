```python
import numpy as np

def apple(Board,k):
    for i in range(0,k):
        a, b = map(int, input('   : ').split())
        if a==1 and b==1:
            print("1,1    !")
            apple(Board,1)
        else:
            Board[a-1,b-1]= 2 ##  2

n = int(input("N :"))
k = int(input("  :"))

Board = np.zeros([n,n])
apple(Board,k)

print(Board)
```

```python
import numpy as np



##     
def apple(Board,k):
    for i in range(0,k):
        a, b = map(int, input('   : ').split())
        if a==1 and b==1:
            print("1,1    !")
            apple(Board,1)
        else:
            Board[a-1,b-1]= 2 ##  2

n = int(input("N :"))
k = int(input("  :"))

Board = np.zeros([n,n])
apple(Board,k)




##   ,  
l = int(input("\n   !"))  #   
turns = {}  #   
for _ in range(l):
    x, c = input().split()
    turns[int(x)] = c


print("\n! \n",Board,)

Board = np.pad(Board, ((1,0),(1,0)), 'constant', constant_values=0)
snake = [(1, 1)]  #  
Board[1][1] = 1
dr, dc = 0, 1  #   
time = 0  #  

print("\n ! \n   \n",Board,)





#  
while True:
    time += 1
    
    nr = snake[-1][0] + dr  ##  
    nc = snake[-1][1] + dc  ##  -> 
    
    #       
    if not (1 <= nr <= n and 1 <= nc <= n) or Board[nr][nc] == 1:
        break
    
    #       
    if Board[nr][nc] == 2:
        snake.append((nr, nc))  ## !!
        Board[nr][nc] = 1  ##  !
    
    #     
    else:
        snake.append((nr, nc))
        Board[nr][nc] = 1
        pr, pc = snake.pop(0)  ## 0  
        Board[pr][pc] = 0
    
    #  
    if time in turns:
        if turns[time] == 'L': #  !
            dr, dc = -dc, dr
        else:
            dr, dc = dc, -dr
            
    print(snake)

#  
print(time)
```
