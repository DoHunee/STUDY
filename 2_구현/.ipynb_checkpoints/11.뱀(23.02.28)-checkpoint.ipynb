```python
import numpy as np

def apple(Board,k):
    for i in range(0,k):
        a, b = map(int, input('숫자 두 개를 입력하세요: ').split())
        if a==1 and b==1:
            print("1,1에는 사과를 추가할 수 없습니다!")
            apple(Board,1)
        else:
            Board[a-1,b-1]= 2 ## 사과는 2

n = int(input("N을 입력하세요:"))
k = int(input("사과의 개수를 입력하세요:"))

Board = np.zeros([n,n])
apple(Board,k)

print(Board)
```

```python
l = int(input("방향 변환 횟수를 입력하세요!"))  # 방향 변환 횟수
turns = {}  # 방향 변환 정보
for _ in range(l):
    x, c = input().split()
    turns[int(x)] = c

turns ##위치 정보
```

```python

Board = np.pad(Board, ((1,0),(1,0)), 'constant', constant_values=0)

snake = [(1, 1)]  # 뱀의 위치
Board[1][1] = 1

dr, dc = 0, 1  # 뱀의 이동 방향
time = 0  # 게임 시간

print(Board)
```

```python
## 뱀 꼬리 머리 

# 게임 실행
while True:
    time += 1
    
    nr = snake[-1][0] + dr  ## 위 아래
    nc = snake[-1][1] + dc  ##  ->   <-
    
    
    # 벽 또는 자신의 몸과 부딪히면 게임 종료
    if not (1 <= nr <= n and 1 <= nc <= n) or Board[nr][nc] == 1:
        break
    
    # 사과가 있으면 뱀의 길이를 늘리고 사과를 없애줌
    if Board[nr][nc] == 2:
        snake.append((nr, nc))
        Board[nr][nc] = 1
    
    # 사과가 없으면 뱀의 길이를 줄임
    else:
        snake.append((nr, nc))
        Board[nr][nc] = 1
        pr, pc = snake.pop(0)
        Board[pr][pc] = 0
    
    # 방향 변환
    if time in turns:
        if turns[time] == 'L':
            dr, dc = -dc, dr
        else:
            dr, dc = dc, -dr

# 결과 출력
print(time)
```

```python
import numpy as np



## 배열 만들고 사과 갯수 입력
def apple(Board,k):
    for i in range(0,k):
        a, b = map(int, input('숫자 두 개를 입력하세요: ').split())
        if a==1 and b==1:
            print("1,1에는 사과를 추가할 수 없습니다!")
            apple(Board,1)
        else:
            Board[a-1,b-1]= 2 ## 사과는 2

n = int(input("N을 입력하세요:"))
k = int(input("사과의 개수를 입력하세요:"))

Board = np.zeros([n,n])
apple(Board,k)




## 방향변환 횟수 , 위치 정보
l = int(input("\n방향 변환 횟수를 입력하세요!"))  # 방향 변환 횟수
turns = {}  # 방향 변환 정보
for _ in range(l):
    x, c = input().split()
    turns[int(x)] = c


print("\n패딩전! \n",Board,)

Board = np.pad(Board, ((1,0),(1,0)), 'constant', constant_values=0)
snake = [(1, 1)]  # 뱀의 위치
Board[1][1] = 1
dr, dc = 0, 1  # 뱀의 이동 방향
time = 0  # 게임 시간

print("\n 패딩후! \n 뱀위치 추가 \n",Board,)





# 게임 실행
while True:
    time += 1
    
    nr = snake[-1][0] + dr  ## 위 아래
    nc = snake[-1][1] + dc  ##  -> 
    
    # 벽 또는 자신의 몸과 부딪히면 게임 종료
    if not (1 <= nr <= n and 1 <= nc <= n) or Board[nr][nc] == 1:
        break
    
    # 사과가 있으면 뱀의 길이를 늘리고 사과를 없애줌
    if Board[nr][nc] == 2:
        snake.append((nr, nc))  ## 추가!!
        Board[nr][nc] = 1  ## 사과 없애!
    
    # 사과가 없으면 뱀의 길이를 줄임
    else:
        snake.append((nr, nc))
        Board[nr][nc] = 1
        pr, pc = snake.pop(0)  ## 0은 당연히 꼬리겠지
        Board[pr][pc] = 0
    
    # 방향 변환
    if time in turns:
        if turns[time] == 'L': # 방향 바꿔주는겨!
            dr, dc = -dc, dr
        else:
            dr, dc = dc, -dr
            
    print(snake)

# 결과 출력
print(time)
```
