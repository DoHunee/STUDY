```python
def combination(arr,n):
    result=[]
    
    if n == 1:
        for i in arr:
            result.append([i])
    
    
    elif n >1:
        for i in range(len(arr) - n+1): 
            for j in combination(arr[i+1:],n-1):
                result.append([arr[i]]+j)
                
    return result

arr = [1,2,3]
print(combination(arr,1))
print(combination(arr,2))
print(combination(arr,3))
```

```python

'''
볼링공의 무게를 쭉 준다
>>고를 수 있는 모든 조합을 구하라 
>> combination

[1,3],[1,3] >> 이런식으로 겹치는건 노상관 
[3,3],[2,2] >> 이런 경우를 제외하쟈!

나는 모든 경우의 수를 배열에 담고
중복되는 경우를 count해서 뺐으


5
3
1 3 2 3 2
>>8

8
5
1 5 4 3 2 4 5 2
>>25


'''
```

```python
## 입력
n = int(input("볼링공 갯수 입력:"))
m = int(input("최대 무게 입력:"))  
arr = list(map(int, input("볼링공 무게 입력").split())) 

if m<max(arr):
    print("최대 무게를 넘어갑니다")
```

```python
## arr = [1,3,2,3,2]  >> 앞쪽 입력 내용

result=[]  ## 나올 수 있는 모든 경우의 수를 담을 배열
k=2 ## 몇개로 묶는지!! >> 여기는 2개 조합만 생각!
cnt=0 ## 중복되는부분 카운트 할 변수

if k == 1:  ## 1개로 묶을거면 그냥 배열에 추가 >>당연!
        for i in arr:
            result.append([i])

elif k >1: ##2개부터 조건을 달아야지
    for i in range(0,len(arr)-1):  
        print(arr[i])
        
        for j in combination(arr[i+1:],1): ## slice를 써야 끝값 넘어가는거 막아준다
            print(j)                       ## 배열끼리 비교해야 하니까 이걸로 함
            
            result.append([arr[i]]+j)  ## result에 삽입
            
            if [arr[i]] == j:   ##중복되는거 count ex) (3,3) , (2,2)이런거
                cnt+=1
                
                
for i in result:
    print(i)

print("\n모든 경우의 수:",len(result))
print("중복 수:",cnt)
print("답:",len(result)-cnt)
       
```
