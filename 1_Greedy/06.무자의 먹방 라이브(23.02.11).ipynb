```python
음식번호랑 음식시간
가장 가까운 번호의 음식!

무지는 1초 먹고 
다음 음식으로 넘어간다!

네트워크 오류가 K초 동안 발생

마지막 음식을 먹는다!
-1을 반환!!

시간이 담겨있는 배열 food times
K초 후

'''

먹기 위해 사용한 시간 + (현재 음식 시간 - 이전 음식 시간) * 

현재 남은 음식 개수와  k를 비교하여 작다면 while문을 코드를 실행한다. 
k보다 작다는 것은 현재 음식을 뺄 수 있다는 말이다. 
현재 음식을 빼기 위해 현재 음식을 큐에서 가져와 now에 저장한다. 
현재 음식을 먹기 위한 시간을 계산하여 
sum_value에 더해준 뒤 다 먹은 음식을 제거하기 위해 length에 -1 해준다. 
현재 음식을 다 먹었으므로 이전 음식을 현재 음식으로 바꿔준다. 
while문을 빠져나왔다면 더 이상 뺄 수 있는 음식이 없으므로 
음식 번호 기준으로 큐를 정렬하여 result에 저장한다. 
마지막으로 다음으로 먹어야 할 음식 번호를 리턴해준다
'''
```

```python
'''
(음식시간,음식번호)

1번 음식: 8초 소요
2번 음식: 6초 소요
3번 음식: 4초 소요
    
큐에 값을 삽입하면 q = [(4,3), (6,2), (8,1)] 이 되고, 
length = 3 이 된다.
k = 15
'''
```

```python
import heapq


def solution(food_times, k):
    
    if sum(food_times) <= k:
        return -1
    
    h = []  # 큐 배열!!!
    for i in range(len(food_times)):
        heapq.heappush(h, (food_times[i], i + 1))  #여기서 i+1은 번호 추가!!
        print(h)

    food_num = len(food_times)  # 남은 음식 개수
    previous = 0  # 이전에 제거한 음식의 food_time

    
    
    
    while h:
        # 먹는데 걸리는 시간: (남은 양) * (남은 음식 개수)
        t = (h[0][0] - previous) * food_num
        # 시간이 남을 경우 현재 음식 빼기
        if k >= t:
            k -= t
            previous, _ = heapq.heappop(h)
            food_num -= 1
        
        # 시간이 부족할 경우(음식을 다 못먹을 경우) 
        #남은 음식 중에 먹어야 할 음식 찾기
        else:
            idx = k % food_num
            h.sort(key=lambda x: x[1])
            answer = h[idx][1]
            break

    return answer
```

```python
'''
(음식시간,음식번호)

1번 음식: 5초 소요
2번 음식: 7초 소요
3번 음식: 4초 소요
    
큐에 값을 삽입하면 q = [(4,3), (5,1), (7,2)] 이 되고, 
length = 3 이 된다.
k = 15
'''
```

```python
food_times = [5,7,2,1,4]   #음식을 먹는데 걸리는 시간배열!
k = 5
solution(food_times, k)
```
