```python
'''
보자보자 어디보자

각자 연속 같은수를 만나는 카운트를 세서
그걸로 뭘 해보까?

변환을 하려면 연속되는게 적은걸 바꿔야 하ㅏ?
중요한건 총 개수가 아니고 변환을 최소화 하는거다!!

'''


'''
1뭉텅이

1 => 0 으로 바뀌는걸 세면 1뭉텅이가 몇개인지 알 수 있고

0뭉텅이 
0 => 1 으로 바뀌는걸 세면 0뭉텅이가 몇개인지 알 수 있지!

아 글고 여기서 마지막 배열에 숫자가 뭔지 보고 카운트 플러스 해준다


ex) 
0100110001     0뭉텅이 - 3   1뭉텅이 - 3 (마지막이 1로끝나니까+1) 
1011001110     0뭉텅이 - 3(마지막이 0으로 끝나니까+1)   1뭉텅이 - 3 
0001100        0뭉텅이 - 2(마지막이 0으로 끝나니까+1)   1뭉텅이 - 1
'''
이래서 더 작은 뭉텅이 횟수가 답이네유~~




```

```python
# num = [0,1,0,0,1,1,0,0,0,1]  # 0-3 1-3
# num = [1,0,1,1,0,0,1,1,1,0]  # 0-3 1-3
# num = [0,0,0,1,1,0,0]  # 0-2 1-1

num= list(map(int, input("0과 1로 이루어진 문자열을 입력하세요!")))


cnt_zero = 0  # 0뭉텅이
cnt_one = 0   # 1뭉텅이


for p1, p2 in zip(num,num[1:]):
    if(p1 < p2):   # 0 > 1
        cnt_zero += 1
    elif(p1 > p2):   
        cnt_one += 1
    
## 배열 마지막 보고 추가
if num[-1] == 0:
    cnt_zero += 1
elif num[-1] == 1:
    cnt_one += 1
    
    

print("0뭉탱이 수:",cnt_zero)
print("1뭉탱이 수:",cnt_one)

if cnt_zero < cnt_one:
    print("최소 출력 예시 = ",cnt_zero)
else:
    print("최소 출력 예시 = ",cnt_one)

    
    

```
