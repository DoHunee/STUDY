```python
##   !
n = int(input("  :"))  ##    
num_list = list(map(int, input(" :").split()))  #  1  !


if  n != len(num_list):
    print("    ")

    


group = 0 #  
cnt= 0   #  


for i in num_list:
    cnt += 1 #    !
    #print(" ",cnt)
    print(" ",cnt)
    if i <= cnt: #    !   
        group+= 1
        cnt = 0

        
print(group)
```
