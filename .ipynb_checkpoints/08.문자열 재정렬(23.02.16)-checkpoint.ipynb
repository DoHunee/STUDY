```python
'''
대문자랑 숫자
오름차순으로 정렬!
그 뒤에 모든 숫자를 더한 값!
'''
```

```python
arr = "K1KA5CB7"


answer = [] #알파벳 담을 배열 
num= 0  # 숫자들 더할꺼!

for i in arr:
    if i.isalpha():
        answer.append(i)

    else:
        num += int(float(i))

answer.sort()              # 알파벳 순서대로!
answer.append(str(num))    #int를 str로 바꿔서 문자열 대입

print(answer)
print(''.join(answer)) # 배열을 문자열로!
```
