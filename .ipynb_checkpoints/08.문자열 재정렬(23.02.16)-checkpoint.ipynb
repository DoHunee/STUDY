```python
arr = "K1KA5CB7"


answer = [] #   
num= 0  #  !

for i in arr:
    if i.isalpha():
        answer.append(i)

    else:
        num += int(float(i))

answer.sort()              #  !
answer.append(str(num))    #int str   

print(answer)
print(''.join(answer)) #  !
```
