```python
# num = [0,1,0,0,1,1,0,0,0,1]  # 0-3 1-3
# num = [1,0,1,1,0,0,1,1,1,0]  # 0-3 1-3
# num = [0,0,0,1,1,0,0]  # 0-2 1-1

num= list(map(int, input("0 1   !")))


cnt_zero = 0  # 0
cnt_one = 0   # 1


for p1, p2 in zip(num,num[1:]):
    if(p1 < p2):   # 0 > 1
        cnt_zero += 1
    elif(p1 > p2):   
        cnt_one += 1
    
##    
if num[-1] == 0:
    cnt_zero += 1
elif num[-1] == 1:
    cnt_one += 1
    
    

print("0 :",cnt_zero)
print("1 :",cnt_one)

if cnt_zero < cnt_one:
    print("   = ",cnt_zero)
else:
    print("   = ",cnt_one)

    
    

```
