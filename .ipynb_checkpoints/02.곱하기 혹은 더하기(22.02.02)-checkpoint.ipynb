```python
'''
1.배열에 수 받고

2. 가장 큰 수를 만들라믄
 뭐야 시빨 0만 더하고 다 곱하면 되는거자낰ㅋㅋㅋ
 아니지 그냥 0만 빼고 다 곱하면 돼
 
 근데 만약에 마이너스도 들어가면 양수로 바꾸는 조건 주고
 


'''

'''
1.0을 빼쟈
2. 

1이 나오면 다음수를 더하쟈
이걸 몰랐넹

 배열로 받고
 1 다음수
 왜냐면 이게 곱하는게 기본이자나
 (1+)

'''
```

```python
## 1일 경우에 그 다음 작은숫자랑 더하고   
## 큰숫자면 곱하는게 이득 


num_list = list(map(int, input("숫자입력하시오!(20개 제한):")))

# 0 없애주고
for i in num_list:
    if i == 0:
        num_list.remove(0)
        
# 남은거 곱해유        
sum = 1         
for i in num_list:
        sum *= i
        if sum > 200000000:
            print("20억이 넘어가유!")
            num_list = [] #초기화

sum

### ㄹㅇ 5분컷
```
