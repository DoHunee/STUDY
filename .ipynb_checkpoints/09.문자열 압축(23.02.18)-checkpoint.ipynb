```python
arr = "aabbaccc"

cnt = 0
for i in range(1,len(arr)+1):  ##1부터 문자열 길이만큼
    result = arr[i-1:i]
    result2 = arr[i:i+1]
    if result == result2:
        print(result , result2)
        cnt +=1

print(cnt)
```

```python
def solution(s):
    if len(s) == 1:  #길이가 1인경우에는 그냥 그대로 리턴@!!
        ans = len(s)
    
    elif len(s) >2:
        ans = min([compress(s, x) for x in range(1, len(s))])
        ##1부터 문자열 길이 -1 까지 돌리면서 copress 실행!
        ## 최솟값만 반환!!
    return ans



def compress(s, num):  ##문자열 압축하는 함수!!
    chunks = [s[x : x + num] for x in range(0, len(s), num)]
    ## chunk에 문자열 잘라서 대입하는 부분!!
    
    cnt = 1
    now = [] ###압축 결과를 담는 배열!!
    
    for i in range(len(chunks) - 1):
        if chunks[i] == chunks[i + 1]:  ##문자 덩어리 i 를 담덩어리랑 비교!
            cnt += 1 ## 같으면 카운트를 하고!!
        
        else: ##중복되지가 않아!!!
            
            if cnt > 1:  #카운트가 2이상이다 즉,중복되는 숫자가 있으면!
                now.append(str(cnt)) #배열에 추가!!!
            now.append(chunks[i])  ## 그 다음에 문자를 삽입해야지!
            cnt = 1  # 압축 종료!!
    
    if cnt > 1: ## cnt가 2이상이면! 결과에 또 추가해야지
        now.append(str(cnt))
    
    now.append(chunks[i + 1]) #마지막 덩어리를 추가해준다!!!
    return len("".join(now))
    now.append(chunks[i + 1])
```

```python
arr = "ababcdcdababcdcd"
solution(arr)
```
