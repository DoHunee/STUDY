```python
def solution(s):
    if len(s) == 1:  # 1   @!!
        ans = len(s)
    
    elif len(s) >2:
        ans = min([compress(s, x) for x in range(1, len(s))])
        ##1   -1   copress !
        ##  !!
    return ans



def compress(s, num):  ##  !!
    chunks = [s[x : x + num] for x in range(0, len(s), num)]
    ## chunk    !!
    
    cnt = 1
    now = [] ###   !!
    
    for i in range(len(chunks) - 1):
        if chunks[i] == chunks[i + 1]:  ##  i   !
            cnt += 1 ##   !!
        
        else: ## !!!
            
            if cnt > 1:  # 2 ,  !
                now.append(str(cnt)) # !!!
            now.append(chunks[i])  ##    !
            cnt = 1  #  !!
    
    if cnt > 1: ## cnt 2!   
        now.append(str(cnt))
    
    now.append(chunks[i + 1]) #  !!!
    return len("".join(now))
    now.append(chunks[i + 1])
```
