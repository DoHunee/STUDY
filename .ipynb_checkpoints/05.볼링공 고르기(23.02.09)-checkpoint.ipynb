```python
def combination(arr,n):
    result=[]
    
    if n == 1:
        for i in arr:
            result.append([i])
    
    
    elif n >1:
        for i in range(len(arr) - n+1): 
            for j in combination(arr[i+1:],n-1):
                result.append([arr[i]]+j)
                
    return result

arr = [1,2,3]
print(combination(arr,1))
print(combination(arr,2))
print(combination(arr,3))
```
